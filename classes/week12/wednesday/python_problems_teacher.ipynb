### Move Zeros

function move_zeros(lst) that in a given list moves 0's to the end of the list.
[1,0,1,2,0,6,0,0,8] -> [1,1,2,6,8,0,0,0,0]

Part A: create and return a new list
Part B: only modify the original list


```
# Part A
def move_zeros(lst):
    new_lst = []
    for l in lst:
        if l != 0:
            new_lst.append(l)
    while len(new_lst) != len(lst):
        new_lst.append(0)
    return new_lst

# Part B
def move_zeros2(lst):
    cur_pos = 0
    for l in lst:
        if l != 0:
            lst[cur_pos] = l
            cur_pos += 1
    while cur_pos < len(lst):
        lst[cur_pos] = 0
        cur_pos += 1
```

### Square Digits

square every digit of a number and concatenate them.
Example: 9119 -> 811181


```
def square_plus(num):
    lst=[]
    while num:
        digit=num%10
        lst.append(digit)
        num//=10
    lst.reverse()
    result=lst[0]**2
    for i in range(1,len(lst)):
        sq=lst[i]**2
        digits=len(str(sq))
        result=result*(10**digits)+sq
    return result
```

### First Non-Repeating Letter

Return first character in string that is not repeated.
Example: 'stress' -> 't'


```
def first_non_repeating_letter(txt):
    for t in txt:
        if txt.count(t)==1:
            return t
    return ''
```

### Lamp Class

Define a Lamp class with color + boolean 'on', toggle_switch and state().


```
class Lamp:
    def __init__(self,color,on=False):
        self.color=color
        self.on=on
    def toggle_switch(self):
        self.on=not self.on
    def state(self):
        return "The lamp is on" if self.on else "The lamp is off"
```

### Commas

Return decimal representation of number with commas.


```
def commas(num):
    s=str(num)
    n=len(s)
    if n<=3:
        return s
    result=""
    i=n
    while i>3:
        result=","+s[i-3:i]+result
        i-=3
    return s[:i]+result
```

### Cartesia Walk

Return True if directions list is exactly 10 steps and returns to start.


```
def cartesia(lst):
    if len(lst)!=10:
        return False
    x=y=0
    for d in lst:
        d=d.upper()
        if d=='N': y+=1
        elif d=='S': y-=1
        elif d=='E': x+=1
        elif d=='W': x-=1
    return x==0 and y==0
```
