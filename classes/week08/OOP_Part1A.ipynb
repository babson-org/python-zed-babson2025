## 1Ô∏è‚É£ What Is Object-Oriented Programming (OOP)?
So far, you‚Äôve written **procedural code** ‚Äî functions that act on data.

OOP, by contrast, **bundles data and behavior together** inside **objects**.

| Concept | Description |
|----------|-------------|
| **Attributes** | Data that describe the object (e.g., color, balance, breed) |
| **Methods** | Actions the object can perform (e.g., drive, deposit, bark) |

Example:
```python
word = 'hello'
print(word.upper())  # .upper() is a method belonging to the string object
```

## 8Ô∏è‚É£ The Pythonic Way ‚Äî `@property`
`@property` makes getters/setters feel like normal attribute access.

```python
class Thermostat:
    def __init__(self, temperature=72):
        self._temperature = temperature

    @property
    def temperature(self):
        return self._temperature

    @temperature.setter
    def temperature(self, new_temp):
        if 40 <= new_temp <= 90:
            print(f'‚úÖ Setting temperature to {new_temp}¬∞F')
            self._temperature = new_temp
        else:
            print('‚ö†Ô∏è Temperature out of range (40‚Äì90 ¬∞F).')

    @temperature.deleter
    def temperature(self):
        print('üßä Temperature reading deleted.')
        del self._temperature
```

## ‚öôÔ∏è Is `@property` Overloading?
It might *look* like we‚Äôre overloading `temperature`, but we‚Äôre not.
Instead, we‚Äôre creating **one managed attribute** with multiple behaviors.

| Operation | Example | What Happens |
|------------|----------|---------------|
| Read | `t.temperature` | Calls getter (`fget`) |
| Write | `t.temperature = 85` | Calls setter (`fset`) |
| Delete | `del t.temperature` | Calls deleter (`fdel`) |

`@property` is really **attribute overriding**, not overloading ‚Äî it replaces normal variable access with controlled behavior.

## üîß What Is a Decorator?
A **decorator** is a *wrapper function* that takes another function, adds extra behavior, and returns a new one.

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print(f'Calling {func.__name__}...')
        result = func(*args, **kwargs)
        print(f'{func.__name__} finished!')
        return result
    return wrapper
```
```python
@my_decorator
def greet():
    print('Hello world!')

greet()  # calls wrapper() instead of greet() directly
```

## üì¶ Understanding `*args` and `**kwargs`
`*args` collects **positional arguments** into a tuple.
`**kwargs` collects **keyword arguments** into a dictionary.

```python
def demo(*args, **kwargs):
    print('args:', args)
    print('kwargs:', kwargs)

demo(1, 2, 3, name='Alice', active=True)
```
Output:
```
args: (1, 2, 3)
kwargs: {'name': 'Alice', 'active': True}
```

## üî© Static and Class Methods
A **static method** is a function inside a class that **does not access or modify** instance (`self`) or class (`cls`) state.

```python
class MathTools:
    @staticmethod
    def add(x, y):
        return x + y
```
```python
print(MathTools.add(2, 3))  # 5
```
Static methods are **utility functions** that belong to a class for organization but don‚Äôt depend on its data.

### üß† Comparison
| Method Type | First Parameter | Access | Use Case |
|--------------|----------------|---------|-----------|
| Instance | `self` | instance data | modify object state |
| Class | `cls` | class data | modify all instances |
| Static | none | no data | helper / conversion |
