## 1Ô∏è‚É£ What Is a Class?
A **class** is a *blueprint* ‚Äî a template that defines what data and behaviors an object will have.

Think of a class like a recipe: it describes *what ingredients* (data) each object will have and *what actions* (methods) it can perform.

```python
class BankAccount:
    pass  # placeholder
```
This creates a new *data type* called `BankAccount`, but no actual data yet.

## 2Ô∏è‚É£ What Is an Object?
An **object** is a specific *instance* of a class ‚Äî a real thing created from that blueprint.

```python
acct = BankAccount()
```
Here, `BankAccount` is the *blueprint* and `acct` is an *object* built from it. Each object has its own data.

Before, in procedural code, we would do `lst = list([1,2,3,4])` or `lst =[1,2,3,4]`.  
`lst` is actually an instance of the list class.

## 3Ô∏è‚É£ What Is `__init__()`?
`__init__` (read ‚Äúdunder init‚Äù) is a **special method** that runs automatically when a new object is created.  
It initializes the object‚Äôs **state** ‚Äî sets up its internal data (attributes).

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
```

When you do:
```python
acct = BankAccount('Alice', 1000)
```
Python automatically calls:
```python
BankAccount.__init__(acct, 'Alice', 1000)
```
`__init__` is often referred to as a **constructor** ‚Äî it constructs or instantiates the object.

## 4Ô∏è‚É£ What Is `self`?
`self` is a reference to *the current object* ‚Äî the one being created or used.  
It tells Python which object‚Äôs data you‚Äôre working with.

```python
self.balance = balance
```
means ‚Äústore the value of `balance` inside *this* object‚Äôs balance attribute.‚Äù  
Each object has its own copy of the data.  
We use the `__init__` function to set up initial values for `self`‚Äôs attributes.

## 5Ô∏è‚É£ Instance Methods and the Role of `self`
Most methods you define are **instance methods** ‚Äî they operate on a specific object and can access or modify that object‚Äôs state.

```python
class BankAccount:
    def deposit(self, amount):
        self.balance += amount
```

### üß† The Rule About `self`
When you **define** an instance method, you must include `self` as the first parameter.  
When you **call** the method, you **omit `self`** ‚Äî Python passes it automatically.

This call:
```python
acct.deposit(100)
```
is equivalent to:
```python
BankAccount.deposit(acct, 100)
```
> So `self` is a conventional name for ‚Äúthe object itself.‚Äù

### ‚öôÔ∏è Example: Seeing `self` in Action
```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def show(self):
        print(f'Account owner: {self.owner}, balance: ${self.balance}')

acct = BankAccount('Alice', 1000)
acct.show()  # Python passes acct as self
```

Output:
```text
Account owner: Alice, balance: $1000
```

### üß≠ Three Kinds of Methods (We‚Äôll Learn 2 Later)
| Method Type | First Parameter | Purpose | Access to Object Data |
|--------------|----------------|----------|------------------------|
| Instance method | `self` | Work on one object | ‚úÖ Yes |
| Class method | `cls` | Work on the class as a whole | ‚öôÔ∏è Later |
| Static method | (none) | Independent helper inside class | ‚öôÔ∏è Later |

### ‚úÖ Quick Summary
| Concept | Meaning |
|----------|----------|
| `self` | Refers to the object itself inside a class |
| Instance method | Function that belongs to an object |
| Definition syntax | `def method_name(self, ...)` |
| Call syntax | `object.method_name(...)` (Python passes `self` automatically) |
| Purpose | Let each object control its own data |

## 6Ô∏è‚É£ Why Do We Need All This?
`class`, `__init__`, and `self` work together to bundle data and behavior so each object manages its own **state**.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print('Insufficient funds')

acct = BankAccount('Alice', 1000)
acct.deposit(200)
acct.withdraw(50)
print(acct.balance)  # 1150
```

## üß© 7Ô∏è‚É£ Procedural vs. Object-Oriented: Putting It All Together

### üß† Procedural Programming
- Functions are separate from data.
- Data must be passed around or stored in global variables.
- Program **state** is shared and difficult to control.

```python
balance = 1000

def deposit(amount):
    global balance
    balance += amount

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount
    else:
        print('Insufficient funds')

deposit(200)
withdraw(50)
print(balance)  # 1150
```

### üß± Object-Oriented Programming
- Data and behavior live together inside **objects**.
- Each object manages its own **state**.
- No global variables needed ‚Äî state is **encapsulated**.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print('Insufficient funds')

acct = BankAccount('Alice', 1000)
acct.deposit(200)
acct.withdraw(50)
print(acct.balance)  # 1150
```

### üìä Comparison Table
| Concept | Procedural | Object-Oriented |
|----------|-------------|----------------|
| Code structure | Functions | Classes & Objects |
| Data location | Global / shared | Inside each object |
| Relationship | Functions *use* data | Objects *own* data |
| State management | Global, fragile | Encapsulated, local |
| Example metaphor | Recipe steps | Chef with ingredients |
| Best for | Small, simple programs | Large, real-world models |

## 9Ô∏è‚É£ Using Classes Across Files

So far, we‚Äôve been defining and using classes in the same notebook or script.  
In real projects, we usually put classes in **separate `.py` files** and then **import** them where needed.

### Example Setup
```
project/
‚îÇ
‚îú‚îÄ‚îÄ bank_account.py
‚îî‚îÄ‚îÄ main.py
```

#### In `bank_account.py`
```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def __str__(self):
        return f"Account({self.owner}, balance={self.balance})"
```

#### In `main.py`
##### Option 1 ‚Äî Import the Class Directly
```python
from bank_account import BankAccount

acct1 = BankAccount("Alice", 1000)
acct1.deposit(200)
print(acct1)
```
‚úÖ Output: `Account(Alice, balance=1200)`

##### Option 2 ‚Äî Import the Whole Module
```python
import bank_account

acct2 = bank_account.BankAccount("Bob", 500)
acct2.deposit(100)
print(acct2)
```
‚úÖ Output: `Account(Bob, balance=600)`

## üîß üîü Project Organization and Best Practices

### üß± General Rule
> Each class should go in its own `.py` file if it represents a distinct concept or major component of your program.

### ‚úÖ Why This Is Good Practice
1. **Readability** ‚Äì Each file is short and focused.  
2. **Maintainability** ‚Äì Fixes stay isolated.  
3. **Reusability** ‚Äì Easy to copy or import into other projects.  
4. **Testing** ‚Äì Unit tests can import one class per file.  
5. **Organization** ‚Äì Logical grouping into folders.

### üóÇÔ∏è Example Structure
```
bank/
‚îú‚îÄ‚îÄ __init__.py
‚îú‚îÄ‚îÄ account.py
‚îú‚îÄ‚îÄ customer.py
‚îî‚îÄ‚îÄ transaction.py
```

Then import like:
```python
from bank.account import BankAccount
from bank.customer import Customer
```

### ‚öôÔ∏è When It‚Äôs Okay to Combine Classes
You can group small, related classes:
```python
# shapes.py
class Rectangle: ...
class Circle: ...
class Triangle: ...
```

### üí¨ Takeaway
> Treat each class like a **building block** ‚Äî give it its own file, import it where needed, and keep each file focused on a single purpose.
