```python
'''
(enlarge sections as needed)
1) Computational thinking 7 1/2 points:

-Abstraction: The long input should be treated as a single string of digits. 6 consecutive digits will become a substring. We have to slide a fixed sized substring across the string and compute the substring's digit product.
- Decomposition: 
    1) Clean and prepare the string by removing newlines and whitespaces
    2) Loop over the starting positions of the 6 digit substrings
    3)Compute the product of the 6 digits
    4) Track the maximum product and track how many times it occurs

    - Pattern Recognoition: Every window is proessed the same way. 6 digits are taken, then multiplied, compared to the current max. This is repeated and is structure as a loop
    - Algorithim Design: Initialize max_prod and count. for each index i from 0 to len(num_str)-6, compute the product of the digits in num_str[i:i+6]. If the products are larger than max_prod, update max_prod and reset count to 1. If it is equal to max_prod, increment count.
    -Scaling: The string is very long but the substring size is fixed at 6. The runtime will then grow linearly with the number of digits.









2) Pseudo code 7 1/2 points:

Declare max_prod and set to -1, count = 0

Remove spaces/newlines from num_str

If length < 6: return (0,0) 

For each starting position in the string where a substring of 6 digits can fit
select the next 6 consecutive digits and set prod to 1

For each digit in the selected substring
convert the character into a number
then multiply it into prod

if prod is greater than max_prod
update max_prod to prod
set count to 1

Else if prod equals max_prod
Incrase the count by 1

Return (max_prod, count)









3) Actual code 10 points:

Read the following as one continuous string:

73167176531330624919225119674426574742355349194934
96983520312774506326239578318016984801869478851843
85861560789112949495459501737958331952853208805511
12540698747158523863050715693290963295227443043557
66896648950445244523161731856403098711121722383113
62229893423380308135336276614282806444486645238749
30358907296290491560440772390713810515859307960866
70172427121883998797908792274921901699720888093776
65727333001053367881220235421809751254540594752243
52584907711670556013604839586446706324415722155397
53697817977846174064955149290862569321978468622482
83972241375657056057490261407972968652414535100474
82166370484403199890008895243450658541227588666881
16427171479924442928230863465674813919123162824586
17866458359124566529476545682848912883142607690042
24219022671055626321111109370544217506941658960408
07198403850962455444362981230987879927244284909188
84580156166097919133875499200524063689912560717606
05886116467109405077541002256983155200055935729725
71636269561882670428252483600823257530420752963450

'''
num_str = (
    "73167176531330624919225119674426574742355349194934"
    "96983520312774506326239578318016984801869478851843"
    "85861560789112949495459501737958331952853208805511"
    "12540698747158523863050715693290963295227443043557"
    "66896648950445244523161731856403098711121722383113"
    "62229893423380308135336276614282806444486645238749"
    "30358907296290491560440772390713810515859307960866"
    "70172427121883998797908792274921901699720888093776"
    "65727333001053367881220235421809751254540594752243"
    "52584907711670556013604839586446706324415722155397"
    "53697817977846174064955149290862569321978468622482"
    "83972241375657056057490261407972968652414535100474"
    "82166370484403199890008895243450658541227588666881"
    "16427171479924442928230863465674813919123162824586"
    "17866458359124566529476545682848912883142607690042"
    "24219022671055626321111109370544217506941658960408"
    "07198403850962455444362981230987879927244284909188"
    "84580156166097919133875499200524063689912560717606"
    "05886116467109405077541002256983155200055935729725"
    "71636269561882670428252483600823257530420752963450"
)


from typing import Tuple

def largest_six_product(num_str: str) -> Tuple[int, int]:
    # Remove spaces and newlines so the string is continuous
    cleaned = "".join(num_str.split())

    # If fewer than 6 digits, return (0, 0)
    if len(cleaned) < 6:
        return 0, 0

    max_prod = -1
    cnt = 0

    # Loop through possible groups of 6 digits
    for i in range(len(cleaned) - 6 + 1):
        window = cleaned[i:i+6]
        prod = 1

        # Multiply the digits in the window
        for ch in window:
            prod *= int(ch)

        # Update max_prod and count
        if prod > max_prod:
            max_prod = prod
            cnt = 1
        elif prod == max_prod:
            cnt += 1

    return max_prod, cnt




print(largest_six_product(num_str))
```
